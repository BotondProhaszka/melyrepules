```python
df = pd.read_csv('data/train_metadata.csv')
df
```

```python
#get the size of a file
train_folder_root = 'data/train_audio/'
def getSize(filename):
    st = os.stat(train_folder_root+filename)
    return st.st_size

#add a column to the dataframe with the size of the file
print('Adding size column to dataframe')
df['size'] = df['filename'].apply(getSize)
```

```python
df
```

```python
#list the duplicated rows in df3 except the filename column
df_duplicated = df[df.duplicated(subset=df.columns.difference(['filename', 'url']))]
print('Number of duplicated rows:', len(df_duplicated))
```

```python
#delete the dulicated rows in df3 but keep the first one
df_cleaned = df.drop_duplicates(subset=df.columns.difference(['filename', 'url']), keep='first')
df_cleaned
```
